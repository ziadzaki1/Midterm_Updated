## What is Pandas?
Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool, built on top of the Python programming language.<br>

##Library Highlights
- A fast and efficient DataFrame object for data manipulation with integrated indexing;

- Tools for reading and writing data between in-memory data structures and different formats: CSV and text files, Microsoft Excel, SQL databases, and the fast HDF5 format;

- Intelligent data alignment and integrated handling of missing data: gain automatic label-based alignment in computations and easily manipulate messy data into an orderly form;

- Flexible reshaping and pivoting of data sets;

- Intelligent label-based slicing, fancy indexing, and subsetting of large data sets;

- Columns can be inserted and deleted from data structures for size mutability;

- Aggregating or transforming data with a powerful group by engine allowing split-apply-combine operations on data sets;

- High performance merging and joining of data sets;

- Hierarchical axis indexing provides an intuitive way of working with high-dimensional data in a lower-dimensional data structure;

- Time series-functionality: date range generation and frequency conversion, moving window statistics, date shifting and lagging. Even create domain-specific time offsets and join time series without losing data;

- Highly optimized for performance, with critical code paths written in Cython or C.

- Python with pandas is in use in a wide variety of academic and commercial domains, including Finance, Neuroscience, Economics, Statistics, Advertising, Web Analytics, and more.

##Mission
Pandas aims to be the fundamental high-level building block for doing practical, real world data analysis in Python. Additionally, it has the broader goal of becoming the most powerful and flexible open source data analysis / manipulation tool available in any language.

```python
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf

# Fetch AAPL data
df_stock = yf.download('AAPL', start='2007-01-01', end='2025-04-28')

df_launch = pd.read_excel("sample_data/launch.xlsx")

df_launch['Date'] = pd.to_datetime(df_launch.date)
df_stock['Date'] = pd.to_datetime(df_stock.index)
df_stock.insert(0, 'ID', range(0, len(df_stock)))
print(df_stock)
print(df_launch)

# Initialize list to store matched IDs
matched_ids = []

# Loop through each launch date
for launch_date in df_launch['Date']:
    # Try matching exact date, +1 day, or +2 days
    for offset in [0, 1, 2]:
        check_date = launch_date + pd.Timedelta(days=offset)
        matching_row = df_stock[df_stock['Date'] == check_date]

        # If a match is found, store the ID and stop checking offsets
        if not matching_row.empty:
            matched_ids.append(int(matching_row['ID'].iloc[0]))
            break
    else:
        # If no match is found after checking all offsets, print a message
        print(f"No trading day found for {launch_date} within 2 days")

print(matched_ids)

plt.figure("Apple Stock")
plt.figure(figsize = (20, 10))
plt.plot(df_stock["Date"], df_stock["Close"], '-^', ms=11, mfc="red", linewidth=0.6,
         markevery=matched_ids, label="Iphone launch date")
plt.xlabel("Dates")
plt.legend(loc="upper left")
plt.show()
df_stock.to_csv("sample_data/result.csv")
```

```python
# Plotting using pandas' built-in functionality
ax = df_stock['Close'].plot(
    figsize=(20, 10),
    style='-ob',
    linewidth=1.5,
    markevery=matched_ids,  # Highlight the launch dates
    markerfacecolor='green',
    markersize=10,
    label='Close Price'
)

# Highlight iPhone launch dates with vertical lines
for launch_date in df_stock.iloc[matched_ids].index:
    ax.axvline(x=launch_date, color='gray', linestyle='--', linewidth=1.5)

# Labeling and legend
ax.set_xlabel("Date")
ax.set_ylabel("Close Price")
ax.set_title("Apple Stock Prices with iPhone Launch Dates Highlighted")
ax.legend(loc="upper left")

# Show the plot
plt.show()
```

```python
from matplotlib import pylab as plt
import pandas as pd
import yfinance as yf

company = "NVDA"
df = yf.download(company, start='2025-01-01', end='2025-04-29')

print(df.columns)
# Ensure single-level columns
if isinstance(df.columns, pd.MultiIndex):
    df.columns = df.columns.get_level_values(0)  # Use the first level (e.g., 'NVDA')

print(df.columns)
plt.style.use("default")

plt.figure(f"{company} CandleStick", figsize=(20,10))
plt.title(f"{company} CandleStick")
plt.vlines(x=df.index, ymin=df.Low, ymax=df.High, color="gray")

# Filter for green and red, and drop NaNs
print(df.Open)
green = df.loc[df.Open <= df.Close, ['Open', 'Close', 'High', 'Low']]
green['Height'] = green.Close - green.Open

red = df.loc[df.Open > df.Close, ['Open', 'Close', 'High', 'Low']]
red['Height'] = red.Open - red.Close

# Debug: Check data
print("Green DataFrame:")
print(green[['Open', 'Close', 'Height']].head())
print(green.dtypes)

print("Red DataFrame:")
print(red[['Open', 'Close', 'Height']].head())
print(red.dtypes)

# Plot with NumPy arrays
plt.bar(x=green.index, height=green.Height, bottom=green.Open, color="green")
plt.bar(x=red.index, height=red.Height, bottom=red.Close, color="red")

plt.show()
```

```python
import pandas as pd

# Load the data
df = pd.read_csv("https://raw.githubusercontent.com/hadley/data-baby-names/master/baby-names.csv")

print(df)
# Filter data according to specific criteria
df_filtered = df[(df.year > 2005) & (df.sex == "boy") & (df.percent > 0.005)]

# Group by 'name' and calculate the mean of the 'percent' column only
grouped = df_filtered.groupby('name')['percent'].mean()
print(grouped)

# Convert the Series back to DataFrame for further manipulation if necessary
mean_df = grouped.reset_index()

# Sort the data by 'percent' in descending order
sorted_mean_df = mean_df.sort_values(by="percent", ascending=False)

# Display the results
print("Final Results")
print(sorted_mean_df.values)
```
